```python
import pandas as pd
import matplotlib.pyplot as plt
from config import TECHNOLOGIES
```

```python
df = pd.read_csv("job_scraper\job_listings.csv")
```

```python
import re

tech_counts = {tech: df["Description"].str.count(tech, flags=re.IGNORECASE).sum() for tech in TECHNOLOGIES}
```

```python
plt.figure(figsize=(20, 8))
plt.bar(tech_counts.keys(), tech_counts.values())
plt.xticks(rotation=90)
plt.xlabel("Technologies")
plt.ylabel("Mentions")
plt.title("Technology Popularity in Job Descriptions")
plt.tight_layout()
plt.savefig("technology_popularity.png")
```
